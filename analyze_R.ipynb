```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
R = np.array(pd.read_csv("R_values.csv",header=None))
eps = np.linspace(0,10,201)
```

```python
numthings = 201
R_array = np.zeros(numthings)

for i in range(numthings):
    R_array[i] = float(R[0][i])
```

```python
plt.figure()
plt.plot(eps, R_array)
plt.xlabel("Energy")
plt.ylabel("R")
plt.title("R vs Energy for np.linspace(0,10,201)")
plt.show()
```
