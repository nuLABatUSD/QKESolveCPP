```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
my_array = np.array(pd.read_csv("energy_density_and_pressure_for_T.csv"))
```

```python
my_array
```

```python
temp_array = my_array[:,0]
density_array = my_array[:,1]
pressure_array = my_array[:,2]
```

```python
plt.figure()
plt.plot(temp_array,density_array,label="energy density")
plt.plot(temp_array,pressure_array,label="pressure")
plt.legend()
plt.show()
```
